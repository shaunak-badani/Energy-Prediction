```python
from __future__ import absolute_import, division, print_function

import tensorflow as tf
from tensorflow.keras import Model, layers
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import mnist
```

* `mnist.load_data` downloads mnist data in format:  <br />
$ (train_{data}, train_{labels}), (test_{data}, test_{labels}) $

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

```python
def np_size(x):
    '''
    Prints memory that a numpy array occupies
    '''
    print("Size of array : ", x.size)
    h = x.size * x.itemsize
    h /= 1024
    print("Memory occupied by array in KiloBytes : ", h)
    print("Memory occupied by array in MegaBytes : ", h / 1024)    
```

```python
def show_images(x, num = 100, caption = None):
    '''
    Args :
    x => (n * 28 * 28 array, containing n images)
    num => number of images to be shown
    caption => Array of size n * 1 containing captions to be put on top of the images
    '''
    assert num % 10 == 0, "Num must be a multiple of 10!"
    rows = num // 10
    fig, axs = plt.subplots(rows, 10, figsize = (10, rows))

    for i in range(rows):
        for j in range(10):
            h = i * 10 + j
            axs[i, j].imshow(x[h], cmap="gray")
            axs[i, j].axis('off')
            axs[i, j].set_title(caption[h])
    plt.tight_layout()
```

```python
n = 50
show_images(train_data, num = n, caption = train_labels[:n])
```

## Building a multi layer perceptron using Tensorflow

>  Step 1 : Prepare MNIST data to be entered into the network

* Convert to float32
* Flatten
* Normalize to [0, 1]

```python
train_data, test_data = train_data.astype('float32'), test_data.astype('float32')
x_train = train_data
x_test = test_data

x_train /= 255
x_test /= 255
```

```python
x_train = tf.data.Dataset.from_tensor_slices((x_train, train_labels))
```

```python
x_train = x_train.repeat().shuffle(5000).batch(256).prefetch(1)
```

```python
optimizer = tf.optimizers.SGD(0.1)
model = keras.Sequential([layers.Flatten(input_shape = (28, 28)), layers.Dense(128, activation = tf.nn.relu), 
                            layers.Dense(256, activation = tf.nn.relu),
                            layers.Dense(10)])
model.build()
```

```python
print(model.summary())
```

```python
for step, (batchx, batchy) in enumerate(x_train.take(2000), 1):
    
    # Run optimization
    with tf.GradientTape() as g:
#         fc1 = layers.Dense(128, activation = tf.nn.relu)
#         batchx = fc1(batchx)
#         fc2 = layers.Dense(256, activation = tf.nn.relu)
#         batchx = fc2(batchx)
#         pred = layers.Dense(10)
        batch_x = model(batchx)
        # Compute entropy loss
        batchy = tf.cast(batchy, tf.int64)
        loss = tf.nn.sparse_softmax_cross_entropy_with_logits(labels = batchy, logits = batch_x)
        loss = tf.reduce_mean(loss)
        
        
    trainable_variables = model.trainable_variables
    gradients = g.gradient(loss, trainable_variables)
    optimizer.apply_gradients(zip(gradients, trainable_variables))
    if step % 100 == 0:
        pred = model(batchx)
        
        # cross entropy loss
        batchy = tf.cast(batchy, tf.int64)
        loss = tf.nn.sparse_softmax_cross_entropy_with_logits(logits = pred, labels = batchy)
        loss = tf.reduce_mean(loss)
        
        # accuracy
        correct_prediction = tf.equal(tf.argmax(pred, 1), tf.cast(batchy, tf.int64))
        acc = tf.reduce_mean(tf.cast(correct_prediction, tf.float32), axis = -1)
        print("step: %i, loss: %f, accuracy: %f" % (step, loss, acc))
    
```

## Testing the network

```python
test_pred = model(test_data)
preds = np.argmax(test_pred.numpy(), axis = 1)
```

```python
print(preds.shape)
show_images(test_data, num = 50, caption = preds)
```

## Neural Network, from scratch

A few things to note : 
* Tensorflow uses automatic differentiation, beyond the scope of the notebook.
* Hence we will manually derive the backprop equations and use normal matrix multiplication to solve our problem.
* However, we will use Tensorflow in order to shuffle and get data in batch sizes

```python
def relu(x):
    '''
    Args : 
    x: (n * batchsize) vector
    
    Returns : 
    Element wise relu
    '''
    return np.maximum(0, x)

def softmax(x):
    '''
    x : num_classes * batch_size
    '''
    j = x - np.max(x, axis = 0)
    h = np.exp(j)
    h /= h.sum(axis = 0)[np.newaxis, :]
    return h

def cross_entropy_loss(pred, labels):
    '''
    pred : (num_classes * batch_size)
    labels : (batch_size)
    
    returns : 
    cross entropy loss for each element
    '''
    one_hot = np.zeros(pred.shape)
    
    o = one_hot.T
    np.put_along_axis(o, labels.reshape(-1, 1), 1, axis = 1)
    one_hot = o.T
    
    rv = -np.log(pred) * one_hot
    rv = rv.sum(axis = 0)
    # rv => 1 * batch_size thing
    return rv
    
    
```

```python
class MNIST_Classifier:
    '''
    Well defined MLP for classifying MNIST data
    
    Architecture : 784 -> 128 + relu -> 256 + relu -> 10 + softmax

    '''
    
    def __init__(self):
        self.W1 = np.ones((128, 784))
        self.b1 = np.ones((128, 1))
        self.f1 = relu
        
        self.W2 = np.ones((256, 128))
        self.b2 = np.ones((256, 1))
        self.f2 = relu
        
        self.W3 = np.ones((10, 256))
        self.b3 = np.ones((10, 1))
        self.f3 = softmax
        
    def forward_pass(self, x, y):
        assert x.shape[1] == 784, "Input should be batch size * 784!"
        self.X1 = x.T # Mathematical format where columns are data values, shape now 784 * batch_size
        self.X2 = self.f1(self.W1 @ self.X1 + self.b1)
        self.X3 = self.f2(self.W2 @ self.X2 + self.b2)
        self.X4 = self.f3(self.W3 @ self.X3 + self.b3)
        
        self.loss = cross_entropy_loss(self.X4, y)
        self.loss_tot = self.loss.mean()
        
    def backward_pass(self):
        pass
        
        
```
